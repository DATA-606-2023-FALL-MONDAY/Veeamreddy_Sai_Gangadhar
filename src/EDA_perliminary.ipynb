```python
# Import the libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read the features data using pandas
features_data = pd.read_csv("https://raw.githubusercontent.com/DATA-606-2023-FALL-MONDAY/Veeramreddy_Sai_Gangadhar/main/data/Features.csv")
features_data
```

```python
features_data.info()
```

```python
# Read the sales data using pandas
sales_data = pd.read_csv("https://raw.githubusercontent.com/DATA-606-2023-FALL-MONDAY/Veeramreddy_Sai_Gangadhar/main/data/sales.csv")
sales_data
```

```python
sales_data.info()
```

```python
# Read the stores data using pandas
stores_data = pd.read_csv("https://raw.githubusercontent.com/DATA-606-2023-FALL-MONDAY/Veeramreddy_Sai_Gangadhar/main/data/stores.csv")
stores_data
```

```python
stores_data.info()
```

```python
## merging the three datasets to bing the final dataframe
df = sales_data.merge(features_data)
df = df.merge(stores_data)
df = df.fillna(0)
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df.info()
```

```python
#top 5 rows of our final dataframe obtained
df.head()
```

```python
# using get dammies in pandas library to convert all the abject values to be a numeric
df_dum=pd.get_dummies(df)
df_dum.info()
```

```python
# use heatmap in seaborn to find out the correlation between the numeric data
plt.figure(figsize=(15,6))
sns.heatmap(df_dum.corr(),annot=True)
```
