# Data Preprocessing 
## Data Integration
Merging the three CVS files

```python
sales.head()
```

- Printing the first 5 lines of the sales data csv file

```python
# Merging Sales with Features based on "Store," "Date" and "IsHoliday" Columns into a new df called mergedStoreandFeatures 
mergedStoreandFeatures=pd.merge(sales,features, on=['Store','Date', 'IsHoliday'], how='left')
mergedStoreandFeatures.head()
```

- Sales and Features csv files have been merged based "Store," "Date" and "IsHoliday" Columns into a new df called mergedStoreandFeature

```python
# Merging Sales with Features based on "Store" Columns into a new df called finalMergedData
finalMergedData=pd.merge(mergedStoreandFeatures,stores, on=['Store'], how='left')
finalMergedData.head()
```

### Replacing missing values in Markdown 1 to 5 with zeros

```python
# Since only markdown columns have missing values, all the NaNs can replaced with 0 for all the five columns
finalMergedData['MarkDown1'] = finalMergedData['MarkDown1'].apply(lambda x: x if not pd.isna(x) else 0)
finalMergedData['MarkDown2'] = finalMergedData['MarkDown2'].apply(lambda x: x if not pd.isna(x) else 0)
finalMergedData['MarkDown3'] = finalMergedData['MarkDown3'].apply(lambda x: x if not pd.isna(x) else 0)
finalMergedData['MarkDown4'] = finalMergedData['MarkDown4'].apply(lambda x: x if not pd.isna(x) else 0)
finalMergedData['MarkDown5'] = finalMergedData['MarkDown5'].apply(lambda x: x if not pd.isna(x) else 0)
```

### One-hot encoding for "Type" Column

```python
filtered_finalMergedData = pd.get_dummies(filtered_finalMergedData, columns=['Type'], prefix='Type')
filtered_finalMergedData.head()
```

-The type column contains categorical data. There are three types of stores in the dataset A, B, and C. To make them easy to process, one-hot encoding is used o represent the categorical variables as numerical values. in the end, three new columns called ype_A, Type_B, and Type_C are introduced into the dataset.

### Boolean encoding for IsHoliday, Type_A, Type_B, and Type_C columns

```python
filtered_finalMergedData['Type_A'] = filtered_finalMergedData['Type_A'].astype(int)
filtered_finalMergedData['Type_B'] = filtered_finalMergedData['Type_B'].astype(int)
filtered_finalMergedData['Type_C'] = filtered_finalMergedData['Type_C'].astype(int)
filtered_finalMergedData['IsHoliday'] = filtered_finalMergedData['IsHoliday'].astype(int)
filtered_finalMergedData.head()
```

- The IsHoliday, Type_A, Type_B, and Type_C column contains Yes and No entries. Boolean encoding is use to convert "Yes" and "No" into binary values, specifically 0s and 1s for easy and accurate processing 

### Converting Dates to  Standard date format as specifed by ISO 8601

```python
filtered_finalMergedData['Date'] = pd.to_datetime(filtered_finalMergedData['Date'], format='%d/%m/%Y')
```

- In ISO 8601 format, the year comes first (YYYY), followed by the month (MM), and then the day (DD), all separated by hyphens (-)

### How Holidays Affect Sales

```python
salePerYear = filtered_finalMergedData.copy()

salePerYear['Year'] = salePerYear['Date'].dt.year
salePerYear['Month'] = salePerYear['Date'].dt.month

# Groupe by 'Year' and 'Month', and calculate the mean weekly sales and whether it's a holiday per group
mean_weekly_sales_per_month = salePerYear.groupby(['Year', 'Month'])['Weekly_Sales'].mean().reset_index()
mean_weekly_sales_per_month['IsHoliday'] = salePerYear.groupby(['Year', 'Month'])['IsHoliday'].any().reset_index()['IsHoliday']

# Create a bar graph
plt.figure(figsize=(12, 6))
sns.barplot(data=mean_weekly_sales_per_month, x="Month", y="Weekly_Sales", hue="IsHoliday", palette="Blues")

# Set labels and title
plt.xlabel('Month')
plt.ylabel('Mean Weekly Sales')
plt.title('Mean Weekly Sales per Month for All Years (Colored by Holiday)')

# Show the plot
plt.tight_layout()
plt.show()
```

- The impact of holidays on monthly sales is not significant. In reality, the cumulative monthly sales for September are slightly lower than those of the neighboring months. Nevertheless, it can be contended that the drop in sales during September can be attributed to the completion of the summer season in certain regions. With the conclusion of summer vacations, individuals may exhibit a tendency to curtail their expenditures on leisure activities and seasonal products.

### Relationship between Unemployment and Weekly Sales

```python
import pandas as pd
from scipy import stats

# Create a copy of the filtered_finalMergedData
data_copy = filtered_finalMergedData.copy()

# Calculate the correlation coefficient and the p-value
correlation, p_value = stats.pearsonr(data_copy['Unemployment'], data_copy['Weekly_Sales'])

print(f'Correlation coefficient: {correlation:.2f}')
print(f'P-value: {p_value:.4f}')
```

- The relationship between unemployment and weekly sales is statistically significant with a -0.03 correlation coefficient and an extremely low p-value of 0.0000. The small negative association between the two variables, as indicated by the negative correlation coefficient, suggests that weekly sales are likely to fall when unemployment levels rise. The low p-value suggests that this link may be statistically significant and that it is unlikely to be the result of random chance. The possible explanation to this is that unemployment tend to diminish consumer buying power, hence is it expected that when it increases, there is a corresponding drop is sales volume.
